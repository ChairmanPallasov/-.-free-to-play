```python
with io.open('../input/mtxzdata/MTXZ.json', 'r', encoding='utf-8-sig') as f:
    data = json.loads(f.read())
    print(data)
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
Genres = df['Genres'].apply(pd.Series)
Genres
```

```python
gg0 = Genres[0].apply(pd.Series)
gg0 = pd.DataFrame(gg0['Name'])
gg0.rename(columns = {'Name': 'Genre 1'})
```

```python
for i in range(1, len(Genres.columns)):
    gg1 = Genres[i].apply(pd.Series)
    gg0[i] = gg1['Name']
```

```python
gg0['Genre'] = gg0[gg0.columns[0:]].apply(
    lambda x: ', '.join(x.dropna().astype(str)),
    axis=1
)
gg0
```

```python
Genres = Genres[0].apply(pd.Series)
Genres
```

```python
Platforms = df['Platforms'].apply(pd.Series)
gg01 = Platforms[0].apply(pd.Series)
gg01 = pd.DataFrame(gg01['Name'])
gg01.rename(columns = {'Name': 'Platform 1'})
for i in range(1, len(Platforms.columns)):
    gg02 = Platforms[i].apply(pd.Series)
    gg01[i] = gg02['Name']
gg01['Platform'] = gg01[gg01.columns[0:]].apply(
    lambda x: ', '.join(x.dropna().astype(str)),
    axis=1
)
gg01
```

```python
Ratings = df['Ratings'].apply(pd.Series)
gg001 = Platforms[0].apply(pd.Series)
gg001 = pd.DataFrame(gg01['Name'])
gg001.rename(columns = {'Name': 'Rating 1'})
for i in range(1, len(Ratings.columns)):
    gg002 = Ratings[i].apply(pd.Series)
    gg001[i] = gg002['Name']
gg001['Rating'] = gg001[gg001.columns[0:]].apply(
    lambda x: ', '.join(x.dropna().astype(str)),
    axis=1
)
gg001
```

```python
df['Genres'] = gg0['Genre']
df['Platforms'] = gg01['Platform']
df['Ratings'] = gg001['Rating']
MTXZ = df
```

```python
Genres_detail = MTXZ['Genres'].str.get_dummies(sep=',')
Platforms_detail = MTXZ['Platforms'].str.get_dummies(sep=',')
Ratings_detail = MTXZ['Ratings'].str.get_dummies(sep=',')
```

```python
MTXZ = pd.concat([MTXZ, Genres_detail, Platforms_detail, Ratings_detail], axis=1)
```

```python
MTXZ
```

```python
MTXZ.to_csv('/kaggle/working/MTXZ_f.csv')
```
